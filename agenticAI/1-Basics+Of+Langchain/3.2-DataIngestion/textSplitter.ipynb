```python
import json
import requests

json_data = requests.get("https://api.smith.langchain.com/openapi.json").json()
json_data
```

```python
from langchain_text_splitters import RecursiveJsonSplitter
```

```python
js_splitter = RecursiveJsonSplitter(max_chunk_size=300)
js_chunks = js_splitter.split_json(json_data)
```

```python
 js_chunks
```

```python
for c in js_chunks:
    print(c)
```

```python
doc_js = js_splitter.create_documents(texts = [json_data])
for d in doc_js[:3]:
    print(d)
```

```python
texts = js_splitter.split_text(json_data)
print(texts[0])
print(texts[1])
print(texts[2])
```
